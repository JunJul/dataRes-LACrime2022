```python
crm_cd_desc_counts = LACrime["Crm_Cd_Desc"].value_counts()
crm_cd_desc_indices = np.array(crm_cd_desc_counts.index.copy())
```

Make the series into an numpy array to calculate the probability that a crime occur.

```python
pro_crm_cd = np.multiply(np.divide(np.array(crm_cd_desc_counts), np.sum(crm_cd_desc_counts)), 100)
pro_ge_1_indices = np.array(np.where(pro_crm_cd > 1)[0])
```

```python
origin_indices = crm_cd_desc_counts[pro_ge_1_indices].index
```

```python
crm_cd_desc_indices[pro_ge_1_indices] = ['BATTERY-SIMPLE ASSAULT', 'VEHICLE-STOLEN', 'BURGLARY FROM VEHICLE',
                                              'BURGLARY', 'THEFT PLAIN', 'THEFT OF IDENTITY',
                                              'VANDALISM - FELONY', 'INTIMATE PARTNER - SIMPLE ASSAULT',
                                              'ASSAULT WITH DEADLY WEAPON',
                                              'THEFT FROM MOTOR VEHICLE', 'ROBBERY',
                                              'VANDALISM - MISDEAMEANOR',
                                              'THEFT-GRAND', 'CRIMINAL THREAT', 'SHOPLIFTING',
                                              'THEFT FROM MOTOR VEHICLE-GRAND', 'TRESPASSING',
                                              'VIOLATION OF RESTRAINING ORDER']
```

The red color emphasizes these crime cases have a greater probability occuring in LA than the green color.

```python
legend_label = sorted(map(str, list(np.arange(1, 19))), key=int)
legend_label = [i + ": " for i in legend_label]
legend_label = [i + j for i, j in zip(legend_label, crm_cd_desc_indices[pro_ge_1_indices])]
```

```python
not_improtant_cases_len = len(crm_cd_desc_indices) - len(pro_ge_1_indices)
important_cases_len = len(pro_ge_1_indices)

pie_colors = ["red"]*important_cases_len + ["lightgreen"]*not_improtant_cases_len
explode = [0.1]*important_cases_len + [0]*not_improtant_cases_len
labels = list(np.arange(1, 19)) + [""]*not_improtant_cases_len

plt.pie(pro_crm_cd, labels=labels,
        colors=pie_colors,
        explode=explode,
        #rotatelabels=90,
        textprops={'fontsize': 6},
        shadow=True)

plt.legend(labels=legend_label, bbox_to_anchor = (1, 1), ncol = 2)
plt.title("The Most Frequent Crime Case in LA", pad=80)
plt.show()
```

```python
frequentCrimeCase_names = list(origin_indices[pro_ge_1_indices])
frequentCrimeCase_df = LACrime[LACrime["Crm_Cd_Desc"].isin(frequentCrimeCase_names)]
frequentCrimeCase_df = frequentCrimeCase_df.reset_index(drop=True)
```

```python
frequentCrimeCase_df.Weapon_Used_Cd = frequentCrimeCase_df.Weapon_Used_Cd.fillna(0)
frequentCrimeCase_df["weapon_used"] = pd.Series(np.where(frequentCrimeCase_df.Weapon_Used_Cd < 1, frequentCrimeCase_df.Weapon_Used_Cd, 1))
```

```python
weapon_pro = frequentCrimeCase_df.groupby("Crm_Cd_Desc")["weapon_used"].agg(np.mean)
weapon_pro.index = ["ASSAULT WITH DEADLY WEAPON", 'BATTERY-SIMPLE ASSAULT', 'BURGLARY',
                    'BURGLARY FROM VEHICLE', 'CRIMINAL THREAT', 'INTIMATE PARTNER - SIMPLE ASSAULT',
                    'ROBBERY', 'SHOPLIFTING', 'THEFT FROM MOTOR VEHICLE-Grand',
                    "THEFT FROM MOTOR VEHICLE-Petty", 'THEFT OF IDENTITY', 'THEFT PLAIN-Petty',
                   'THEFT-GRAND', 'TRESPASSING', 'VANDALISM - FELONY', 'VANDALISM - MISDEAMEANOR',
                    'VEHICLE-STOLEN', 'VIOLATION OF RESTRAINING ORDER']

colors = ["salmon", "salmon", "lightblue", "lightblue", "salmon", "salmon","salmon"] + ["lightblue"]*11
sns.barplot(x=weapon_pro, y=weapon_pro.index, palette=colors)
plt.xticks(rotation=85)
plt.title("The Probability of Weapon Used in each Crime Case")
plt.xlabel("Weapon Used Probability")
plt.ylabel("Crime Cases")
plt.show()
```

The salmon color emphasizes that the crime cases have an extremely high probability reported with weapons used.
